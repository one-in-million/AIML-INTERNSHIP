### 4) Data Cleaning

#### 4 major steps involved in Data Cleaning
##### 1) Check for Duplicate values
##### 2) Check for null values
##### 3) Check for Data types
##### 4) Remove the unwanted row or column

```python
#Checking for duplicate values
#this one gives the individual result
df.duplicated()
```

```python
#this will give u the overall result and there is no duplicate values present
df.duplicated().any()
```

```python
df.info()
```

```python
#cabin column is not needed so drop
df = df.drop(columns = ['cabin'])
df.info()
```

```python
#replace the missing values using central tendencies
age_mean = df['age'].mean()
fare_mean = df['fare'].mean()
emb_mode = df['embarked'].mode()[0]
df['age'] = df['age'].fillna(age_mean)
df['fare']=df['fare'].fillna(fare_mean)  
df['embarked']=df['embarked'].fillna(emb_mode)  
df.info()
```

### 5) Data Transformation/Feature Engineering

```python
df['age']=df['age'].round(1)   
df['fare']=df['fare'].round(1)
```

```python
df.head(3)
```

```python
df['family_cnt']=df['sibsp']+df['parch']
df=df.drop(columns=['sibsp','parch'])
```

```python
df.head(3)
```

```python
df['embarked'].unique()
df['embarked']=df['embarked'].replace({'C':'Cherbourg','Q':'Queenstown','S':'Southampton'})
```

```python
df = df.rename(columns = {'sex':'gender'})
df
```

```python
df['gender'] = df['gender'].str.title()
df
```

```python
df['name'].str.split(',',n=1,expand=True)
```

```python
#splitting name into 3 parts
df[['Last_name', 'rest']] = df['name'].str.split(',', n=1, expand=True)
df['Last_name'] = df['Last_name'].str.strip()
df['rest'] = df['rest'].str.strip()
df[['Title', 'First_Name']] = df['rest'].str.split(' ', n=1, expand=True)
df['Title'] = df['Title'].str.replace(r'\.$', '', regex=True)
df = df.drop(columns='rest')
df
```

```python
g=df['ticket'].value_counts()
g
```

```python
#creating a new col based on some condition
g=df['ticket'].value_counts()
solo_ticket=g[g==1].index
def f1(d):
    if d['family_cnt']>0 :
        return 'Family'
    else:
        if d['ticket'] in (solo_ticket):
            return 'solo'
        else:
         return "Friends"
```

```python
df['Companion_type']=df.apply(f1,axis=1)
df
```

```python
lst=['ticket','Title','First_Name','Last_name', 'gender','age','Companion_type','family_cnt','pclass','embarked', 'fare','survived']
df=df[lst]
df
```

```python
df['age_cat']=pd.cut(df['age'],bins=[0,15,25,60,81],labels=['child','teen','adult','senior'])
df
```

```python
df.describe()
```

16. Plot the distribution of fare

```python
plt.figure(figsize=(3,4))
a = sns.histplot(df['fare'],bins=10)
plt.show()
```

*Observation :- The distribution is highly right-skewed — most passengers paid cheap fares, and only a small number paid very expensive fares.*

How to Find Right skew or left skew??
Why is it right-skewed even though bars are on the left?

In the histogram, most passengers are indeed piled up on the left (low fares).

But the tail (the long stretch of small counts at higher fares) extends to the right side.

That’s the key:

If the tail goes to the right, it’s right-skewed (positive skew).

If the tail goes to the left, it’s left-skewed (negative skew).

So here, even though most values are on the left (cheap tickets), the rare expensive tickets stretch the distribution to the right → right skew. ✅

df['fare'].skew()

If result > 0 → Right-skewed (tail to the right).

If result < 0 → Left-skewed (tail to the left).

If result ≈ 0 → Symmetric / normal distribution.
Mean > Median → Right-skewed.

Mean < Median → Left-skewed.

Mean ≈ Median → Symmetric.

### Conclusion

Gender played the biggest role → Females had a much higher survival rate than males across all classes and age groups.

Class strongly influenced survival → 1st-class passengers (especially females) had the best chances, while 3rd-class passengers faced the highest risk.

Age mattered → Children had the highest survival rates, whereas adults and seniors were far less likely to survive.

Companion type impacted survival → Passengers traveling with family had better survival chances compared to solo travelers.

Fare distribution was right-skewed → Most passengers paid low fares, but a few very high fares pulled the average upward, and higher fares were linked with better survival.

Embarkation differences existed → Passengers from Cherbourg had higher survival chances, while those from Southampton had the lowest.

In short: Being a female, child, 1st-class, with family, and paying a higher fare greatly increased survival chances on the Titanic.
